# ACTIVE CASES AND CURED CASES VS DEATH CASES

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
x=datai[['new_active', 'new_cured']]
y=datai['death']
x=pd.get_dummies(x)
xt,xte,yt,yte=train_test_split(x,y,test_size=0.3)
md=DecisionTreeRegressor()
md.fit(xt,yt)
print(md.score(xte,yte))
pred=md.predict(xt)
```

```python
datai.columns
```

```python
plt.plot(xt,pred)
plt.xlabel('ACTIVE CASES AND CURED CASES')
plt.ylabel('DEATH CASES')

plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x=datai[['new_active', 'new_cured']]
y=datai['death']
x=pd.get_dummies(x)
xt,xte,yt,yte=train_test_split(x,y,test_size=0.3)
md=LinearRegression()
md.fit(xt,yt)
print(md.score(xte,yte))
pred=md.predict(xt)
```

```python
plt.plot(xt,pred)
plt.xlabel('ACTIVE CASES AND CURED CASES')
plt.ylabel('DEATH CASES')
plt.legend()
```
